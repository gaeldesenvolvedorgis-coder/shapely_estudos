```python
import geopandas as gpd
```

Importação dos vertices dos poligonos das divisões administrativas de Cuba

```python
ponto = "/content/drive/MyDrive/Colab/Dados/cuba_vertices.geojson"
ponto_gpd = gpd.read_file(ponto)
len(ponto_gpd)
```

Limites administrativos em linha de Cuba

```python
linha = "/content/drive/MyDrive/Colab/Dados/cuba_adminlines.geojson"
linha_gpd = gpd.read_file(linha)
linha_gpd.plot()
len(linha_gpd)
```

Selecionar apenas as linhas que tocam os pontos dos vertices.

```python
linhas_sel = linha_gpd[linha_gpd.intersects(ponto_gpd.unary_union)]
len(linhas_sel)
```

O .unary_union transforma todas as geometrias da camada em uma única geometria combinada antes do teste espacial.
cada ponto  VS  geometria única formada por todas as linhas
