```python
from shapely.geometry import Point,Polygon
import geopandas as gpd
```

Enunciado Criar GeoDataFrame

Você possui:

um lago (Polygon)

5 pontos representando casas

```python
gdf = gpd.GeoDataFrame({
    'nome':['casa1','casa2','casa3','casa4','casa5','lago'],
    'geometry':[
        Point(2,1),
        Point(1,3),
        Point(3,2),
        Point(3,5),
        Point(2,3),
        Polygon([(1,2),(1,4),(4,4),(4,2)])]
  }, crs="epsg:4326")
```


Descubra:

quais casas estão dentro do lago

quais estão fora

Para filtrar apenas pontos:

```python
casas =gdf[gdf.geometry.geom_type == 'Point']
lago =gdf[gdf.geometry.geom_type == 'Polygon']
```

```python
#iloc é uma forma de acessar linhas e colunas pela posição numérica no pandas/GeoPandas.
#loc → usa o índice e iloc → usa posição
#Nesse caso se não uso o iloc[] ele entende a feição ainda na posição 5 e dá erro

casas_dentro = casas[casas.intersects(lago.geometry.iloc[0])]
len(casas_dentro['geometry'])
```
