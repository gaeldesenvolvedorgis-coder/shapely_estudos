```python
from shapely.geometry import Point
import geopandas as gpd
```

1. Crie os pontos

```python
p1=Point(2,5)
p2=Point(2,6)
p3=Point(2,7)
p4=Point(2,8)
p5=Point(2,9)
```

2. Monte o GeoDataFrame

```python
gdf = gpd.GeoDataFrame({
        'nome':['Cidade A','Cidade B','Cidade C', 'Cidade D', 'Cidade E'],
        'populacao': [120000,100000,150000,180000,200000],
        'altitude': [450, 370, 485, 257 ,356],
        'geometry':[p1,p2,p3,p4,p5]
    },crs='EPSG:4326')
```

3. Faça consultas

```python
gdf[gdf['populacao'] >= 150000]
```

```python
gdf[gdf['altitude'] >= 400]
```

```python
gdf[gdf['nome'] == 'Cidade A']
```

4. Crie uma nova coluna

```python
gdf['densidade'] = (gdf['populacao']/gdf['altitude'])
```

```python
gdf[gdf['densidade'] <= 300]
```
